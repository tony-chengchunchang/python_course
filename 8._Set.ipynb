## 宣告

- ### 用{}宣告

```python
my_set = {1,2,3}
my_set
```

- ### 型態轉換 set()


    - 將一序列轉換為set

```python
my_set = set([1,2,3])
my_set
```

- ### 可修改，無序值不可重複

    
    - 宣告時會將重複值排除

```python
my_set = {1,1,2,2}
my_set
```

```python
my_set = set([1,1,2,2])
my_set
```

## 集合方法

- ### set.add()


    - 新增元素
    - 必須是不重複值

```python
my_set = {1,2}
my_set.add(3)
my_set
```

- ### set.remove()


    - 移除元素

```python
my_set.remove(3)
my_set
```

- ### 集合概述

    - 交集
    - 聯集
    - 差集

- ### set.intersection() 交集

```python
set1 = {1,2,3}
set2 = {2,3,4}
```

```python
set1.intersection(set2)
```

```python
set1 & set2
```

- ### set.union() 聯集

```python
set1.union(set2)
```

```python
set1 | set2 
```

- ### set.difference() 差集

```python
set1.difference(set2)
```

```python
set1 - set2
```

批新貨商品優惠資訊如下

buy_one_get_one = ['Apple', 'Samsung', 'Mi', 'Asus', 'Acer', 'HP', 'Sony']

free_shipping = ['Apple', 'Samsung', 'Nike', 'UA', 'Mi', 'Asus', 'Uniqlo']

同時做買一送一且免運的商品?

只做免運的商品?

```python
buy_one_get_one = set(['Apple', 'Samsung', 'Mi', 'Asus', 'Acer', 'HP', 'Sony'])
free_shipping =set(['Apple', 'Samsung', 'Nike', 'UA', 'Mi', 'Asus', 'Uniqlo'])
buy_one_get_one.intersection(free_shipping)
buy_one_get_one & free_shipping
free_shipping.difference(buy_one_get_one)
free_shipping - buy_one_get_one
```
